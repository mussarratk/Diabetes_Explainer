```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
plt.figure(figsize=(16, 6))
sns.heatmap(df.corr(),cmap='rocket_r',annot=True) #crest
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.heatmap(df.corr(),cmap='mako_r',annot=True) #crest
plt.show()
```

cmap='rocket_r': The reversed rocket colormap — gives you dark red for high positive correlation, light yellow for lower.

You can also try other colormaps like 'mako_r', 'inferno', or 'viridis_r' if you want subtle or smooth gradients.

vmin=-1, vmax=1 ensures the color scale spans the full possible correlation range.
